```python
%matplotlib notebook
# This import registers the 3D projection, but is otherwise unused.
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401 unused import
import matplotlib.pyplot as plt
import numpy as np

cycle = lambda x, n: np.hstack((x[-n:], x[:-n]))

nCycle = 10
nPoint = 500 + nCycle
A = 0.1*np.random.normal(0,1, nPoint) + 0.5*(np.random.uniform(0,1, nPoint) > 0.95).astype(float)
B = cycle(A, nCycle) + 0.05 * np.random.normal(0,1,nPoint)
x = A[:-nCycle]
y = B[:-nCycle]
z = B[nCycle:]

# 2D plots
plt.figure()
plt.plot(A[:100]+1)
plt.plot(B[:100])
plt.savefig('te_tutorial_traces_good.pdf')

# 3D plots
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.view_init(30, -135)
ax.plot(x, z, 'g.', alpha=0.4, zdir='y', zs=1.5)
ax.plot(y, z, 'r.', alpha=0.4, zdir='x', zs=1.5)
ax.scatter(x,y,z)   # 3D plot

ax.set_xlabel('A(t)')
ax.set_ylabel('B(t)')
ax.set_zlabel('B(t+dt)')
ax.xaxis.set_ticklabels([])
ax.yaxis.set_ticklabels([])
ax.zaxis.set_ticklabels([])

ax.set_xlim([-0.3, 1.5])
ax.set_ylim([-0.3, 1.5])
ax.set_zlim([-0.4, 0.8])

plt.savefig('te_tutorial_scatter_good.pdf') 
plt.show()
```

```python
nCycle = 10
nPoint = 500 + nCycle
A = np.random.normal(0,1, nPoint)
for i in range(1, nPoint):
    A[i] = A[i] + A[i-1]
A -= np.min(A)
A /= np.max(A)
B = A + 0.05 * np.random.normal(0,1,nPoint)
x = A[:-nCycle]
y = B[:-nCycle]
z = B[nCycle:]

# 2D plots
plt.figure()
plt.plot(A[-100:]+1)
plt.plot(B[-100:])
plt.savefig('te_tutorial_traces_bad.pdf')

# 3D plots
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.view_init(30, -135)
ax.plot(x, z, 'g.', alpha=0.4, zdir='y', zs=1.5)
ax.plot(y, z, 'r.', alpha=0.4, zdir='x', zs=1.5)
ax.scatter(x,y,z)   # 3D plot

ax.set_xlabel('A(t)')
ax.set_ylabel('B(t)')
ax.set_zlabel('B(t+dt)')
# ax.xaxis.set_ticklabels([])
# ax.yaxis.set_ticklabels([])
# ax.zaxis.set_ticklabels([])

ax.set_xlim([-0.1, 1.5])
ax.set_ylim([-0.1, 1.5])
ax.set_zlim([-0.3, 1.1])

plt.show()
plt.savefig('te_tutorial_scatter_bad.pdf')
```
