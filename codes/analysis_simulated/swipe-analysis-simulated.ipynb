```python
# Get simulated H5 with TE
te_h5_fnames = gui_fnames("Select files with TE data", "./", "HDF5 (*.h5)")
```

```python
# Extract data
data_dict = {}
for fname in te_h5_fnames:
    with h5py.File(fname, "r") as h5f:
        dataname = os.path.basename(fname[:-3])
        data_dict[dataname] = np.array(list(h5f['results'].values()))
```

```python
P_THR = 0.01 # p-value threshold

# Process
for dataname, data in data_dict.items():
    te, lag, p = data
    nan_mask = np.isnan(p)
    p_nonan = np.copy(p)
    p_nonan[nan_mask] = 100
    pval_mask = p_nonan <= P_THR
    
    print("for", dataname, "mean NCONN is", np.mean(pval_mask))
```
